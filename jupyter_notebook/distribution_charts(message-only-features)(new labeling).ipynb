```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from collections import Counter
```

```python
data = pd.read_csv('../data/data_msg_only(new tags).csv')
```

```python
data_good = data[data.label == 'good']
data_neutral = data[data.label == 'neutral']
data_bad = data[data.label == 'bad']

print(data_good.shape)
print(data_neutral.shape)
print(data_bad.shape)
```

```python
def drow_histogram(x, bins, title):
    plt.hist(x,bins)
    plt.xlabel('Chrarcters Count')
    plt.ylabel('Number of Commits')
    plt.xticks(np.arange(start=0, stop=300, step=30))
    plt.axvline(data_good['characters_count'].mean(), color='r', label='Mean')
    plt.legend()

    plt.title(title)
    plt.tight_layout()
    plt.show()
```

```python
def drow_histogram_2(x, title):
    plt.xlabel('Pattern met')
    plt.ylabel('Number of Commits')
    plt.axvline(x.mean(), color='r', label='mean')
    plt.legend()

    labels, counts = np.unique(np.array(x), return_counts=True)
    plt.title(title)

    plt.bar(labels, counts, align='center')
    plt.gca().set_xticks([0, 0.5, 1])

    plt.tight_layout()
    plt.show()
```

```python
x1 = data_good['characters_count']
x2 = data_neutral['characters_count']
x3 = data_bad['characters_count']
```

```python
drow_histogram(x1, 50, 'Good')
drow_histogram(x2, 50, 'Neutral')
drow_histogram(x3, 200, 'Bad')
```

```python
x1 = data_good['verb_direct_obj']
x2 = data_neutral['verb_direct_obj']
x3 = data_bad['verb_direct_obj']

drow_histogram_2(x1, 'Good')
drow_histogram_2(x2, 'Neutral')
drow_histogram_2(x3, 'Bad')
```

```python
plt.hist(x1, 50, alpha=0.5, label='Good')
plt.hist(x2, 50, alpha=0.5, label='Neutral', color = "tab:green")
plt.hist(x3, 250, alpha=0.5, label='Bad')
plt.legend(loc='upper right')
plt.show()
```
