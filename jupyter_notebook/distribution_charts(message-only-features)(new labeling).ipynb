```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from collections import Counter
```

```python
data = pd.read_csv('../data/data_msg_only(new tags).csv')
```

```python
data_good = data[data.label == 'good']
data_neutral = data[data.label == 'neutral']
data_bad = data[data.label == 'bad']

print(data_good.shape)
print(data_neutral.shape)
print(data_bad.shape)
```

```python
plt.rcParams.update({'font.size': 10})
```

```python
def drow_histogram(x, bins, title):
    plt.hist(x,bins)
    plt.xlabel('Characters Count')
    plt.ylabel('Number of Commits')
    plt.xticks(np.arange(start=0, stop=210, step=30))
    plt.yticks(np.arange(start=0, stop=19, step=2.5))
    plt.axvline(x.mean(), color='r', label='Mean')
    plt.legend()

    plt.title(title)
    plt.tight_layout()
    plt.show()
```

```python
def drow_histogram_2(x, title, x_title):
    plt.xlabel(x_title)
    plt.ylabel('Number of Commits')
    plt.axvline(x.mean(), color='r', label='mean')
    plt.legend()

    labels, counts = np.unique(np.array(x), return_counts=True)
    plt.title(title)

    plt.bar(labels, counts, align='center')
    plt.gca().set_xticks([0, 0.5, 1])
    plt.gca().set_yticks([0, 100, 200, 300, 400])

    plt.tight_layout()
    plt.show()
```

```python
x1 = data_good['characters_count']
x2 = data_neutral['characters_count']
x3 = data_bad['characters_count']
```

```python
drow_histogram(x1, 50, 'Good')
drow_histogram(x2, 50, 'Neutral')
drow_histogram(x3, 200, 'Bad')
```

```python
x1 = data_good['verb_obj_conn']
x2 = data_neutral['verb_obj_conn']
x3 = data_bad['verb_obj_conn']

plt.rcParams.update({'font.size': 12})

drow_histogram_2(x1, 'Good', 'Meet the Pattern')
drow_histogram_2(x2, 'Neutral', 'Meet the Pattern')
drow_histogram_2(x3, 'Bad', 'Meet the Pattern')
```

```python
plt.hist(x1, 50, alpha=0.5, label='Good')
plt.hist(x2, 50, alpha=0.5, label='Neutral', color = "tab:green")
plt.hist(x3, 250, alpha=0.5, label='Bad')
plt.xlabel('Characters Count')
plt.ylabel('Number of Commits')
plt.legend(loc='upper right')
plt.show()
```
