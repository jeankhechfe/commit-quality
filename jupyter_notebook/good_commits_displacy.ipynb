```python
import spacy
from spacy import displacy
```

```python
nlp = spacy.load("en_core_web_lg")
```

```python
nlp.tokenizer.add_special_case('fix', [{ORTH: 'fixing'}])  # treat fix as verb
subject_line = "Switch from pgp.mit.edu to pool.sks-keyservers.net"
subject_line = subject_line.replace(' .', ' ')
subject_line = subject_line.replace('\'', '')
doc1 = nlp(subject_line)
print([(token.text, token.tag_, token.pos_) for token in doc1])
displacy.render(doc1)
```
