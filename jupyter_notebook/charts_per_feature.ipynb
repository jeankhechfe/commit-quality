```python
import pandas as pd
from matplotlib import pyplot as plt
from collections import Counter
```

```python
data = pd.read_csv('../data/data_msg_only(new tags).csv')
x = data.drop(columns=['label'])
y = data['label']
```

```python
data_good = data[data.label == 'good']
data_neutral = data[data.label == 'neutral']
data_bad = data[data.label == 'bad']

print(data_good.shape)
print(data_neutral.shape)
print(data_bad.shape)
```

```python
def drow_pie(x, title):
    slices = [Counter(x)[0], Counter(x)[1]]
    labels = ['Meet the pattern', 'Don\'t meet\nthe pattern']
    colors = ['tab:orange', 'tab:blue']

    plt.pie(slices, labels=labels, colors=colors, autopct='%1.2f%%')

    plt.title(title)
    plt.tight_layout()
    plt.show()
```
